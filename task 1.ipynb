Step 1 : Importing the libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Step 2 : Loading the dataset 

The dataset used is 'Financial Sample Data'

The link for the dataset : https://www.kaggle.com/datasets/luckyrao/financial-sample-data

```python
df = pd.read_excel(r"D:\mit\sem 2\data_viz_lab\Financial Sample.xlsx")
```

```python
df.head()
```

Step 3 : Visualising the categorical data

```python
#1. Histogram 
sns.histplot(df, x = "Product")
```

```python
#2. Bar chart, segment-wise sold units

sns.barplot(data = df, x = 'Segment', y = 'Units Sold')
```

```python
#3. Bar chart - Segment-wise Profit. 
sns.barplot(data = df, x = 'Segment', y = 'Profit')
```

Here, it can be seen that the 'Enterprise' sector shows a Loss.

```python
#4. Bar Chart - Product-wise profit

sns.barplot(data = df, x = 'Product', y = 'Profit')
```

```python
#5. Bar Chart - Country-wise Profit

grp = df.groupby('Country')['Profit'].sum()
sns.barplot(df, x = "Profit", y = "Country", orient = 'h')
```
